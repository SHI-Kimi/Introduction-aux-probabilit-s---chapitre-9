```python
#Exercice 9.9
import timeit

#Code pour préparer l'environnement
code_setup = """
#Fonction proposée à la page 376
import numpy as np
def Bern(p):
    v_a_x = np.random.random()
    if v_a_x <= p:
        return 1
    else:
        return 0

def Binom(n,p):
    somme = 0
    for i in range(n):
        somme += Bern(p)
    return somme

liste_valeurs_n = [2,10,50]
liste_valeurs_p = [0.2,0.7]
"""

code_principal_1 = """
for nombre in liste_valeurs_n:
    for proba in liste_valeurs_p:
        Binom(nombre,proba)
"""

code_principal_2 = """
for nombre in liste_valeurs_n:
    for proba in liste_valeurs_p:
        np.random.binomial(nombre,proba)
"""

fonction_personnalisee = timeit.timeit(stmt=code_principal_1,setup=code_setup,number=10**6)
fonction_binomiale_python = timeit.timeit(stmt=code_principal_2,setup=code_setup,number=10**6)

print("temps fonction personnalisée : {}s".format(int(fonction_personnalisee)))
print("temps fonction python : {}s".format(int(fonction_binomiale_python)))
#Résultat : la fonction que l'on a créée nous-mêmes prend 10x plus de temps que celle de Python (40s vs 4s)
```
